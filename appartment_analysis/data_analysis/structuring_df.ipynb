# Restructure the collected Data

Loading the Scraped Data

```python
import pandas as pd
import json
import numpy as np

# Read JSON data into a variable
json_data = open('../data/scraped/appart_info.json', encoding='utf-8').read() # für eine Python-Datei müsste man noch den Ordner angeben
# Parse JSON into a Python dictionary
data = json.loads(json_data)
# Create DataFrame from the JSON data
df = pd.DataFrame.from_dict(data, orient='index')

# Reset pandas options
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
# Display the DataFrame
df
```

## Clean Baujahr


```python
import re
# Neue Spalte 'Baujahr' hinzufügen
if 'Baujahr' not in df.columns:
    df['Baujahr'] = np.nan


for col in df.columns:
    if col.startswith("Baujahr"):
        split_col = col.split(' ')
        if len(split_col) >= 2:
            year = split_col[1]
            df.loc[df[col] == 1, 'Baujahr'] = float(year)
    elif str('Baujahr') in col:
        # Regex-structure for a construction year between 1700 and 2024
        pattern = r'\b(17[0-9]{2}|20[0-2][0-9]|202[0-4])\b'

        # Extract the construction year from the column name
        year = re.findall(pattern, col)
        if year:
            df.loc[df[col] == 1, 'Baujahr'] = float(year[0])

# Drop columns that start with 'Baujahr' and are not 'Baujahr'
columns_to_drop = [col for col in df.columns if col.startswith("Baujahr") and col != 'Baujahr']
df.drop(columns=columns_to_drop, inplace=True)

df.tail(5)
```

## Verbrauchsausweis

```python
# Neue Spalte 'Verbrauchsausweis' hinzufügen
if 'Verbrauchsausweis' not in df.columns:    
    df['Verbrauchsausweis'] = ''

for col in df.columns:
    if col.startswith("V:"):
        split_col = col.split(' ')
        if len(split_col) >= 3:
            value = split_col[1] + ' ' + split_col[2]
            df.loc[df[col] == 1, 'Verbrauchsausweis'] = str(value)

# Drop columns 
columns_to_drop = [col for col in df.columns if col.startswith("V:") ]
df.drop(columns=columns_to_drop, inplace=True)

df.tail(10)
```

## WLAN

```python
import re
# Neue Spalte 'WLAN' hinzufügen
if 'WLAN_geschw' not in df.columns:   
    df['WLAN_geschw'] = ''



for col in df.columns:
    if str("Mbit/s") in col:
        col_name = col.replace('\n', '').strip()
        col_name = re.sub(r'\s+', ' ', col)
        split_col = col_name.split(' ')
        if len(split_col) >= 3:
            value = split_col[1:]
            value = ' '.join(value)
            df.loc[df[col] == 1, 'WLAN_geschw'] = str(value)

# Drop columns 
columns_to_drop = [col for col in df.columns if "Mbit/s" in col]
df.drop(columns=columns_to_drop, inplace=True)

df.tail(10)
```

## öffentlicher Nahverkehr

```python
import re
# Neue Spalte zur Entfernung bis zum nächsten öffentlichen Verkehrmittel hinzufügen
if 'Entfern_bis_Öff_Nah' not in df.columns:   
    df['Entfern_bis_Öff_Nah'] = ''



for col in df.columns:
    if str("zu Fuß entfernt") in col:
        col_name = col.replace('\n', '').strip()
        col_name = re.sub(r'\s+', ' ', col)
        split_col = col_name.split(' ')
        if len(split_col) >= 3:
            value = split_col[0]
            value = ' '.join(value)
            df.loc[df[col] == 1, 'Entfern_bis_Öff_Nah'] = str(value)

# Drop columns 
columns_to_drop = [col for col in df.columns if "zu Fuß entfernt" in col]
df.drop(columns=columns_to_drop, inplace=True)
df
```

## Geschoss

```python
# Neue Spalte 'Verbrauchsausweis' hinzufügen
if 'Geschoss' not in df.columns:    
    df['Geschoss'] = ''

for col in df.columns:
    if 'OG' in col or 'EG' in col or 'UG' in col or 'Hochparterre' in col or 'Dachgeschoss' in col or 'Souterrain' in col or 'Penthouse' in col or 'Tiefparterre' in col:
        df.loc[df[col] == 1, 'Geschoss'] = col

# Drop columns 
columns_to_drop = [col for col in df.columns if 'OG' in col or 'EG' in col or 'UG' in col or 'Hochparterre' in col or 'Dachgeschoss' in col or 'Souterrain' in col or 'Penthouse' in col or 'Tiefparterre' in col ]
df.drop(columns=columns_to_drop, inplace=True)

df.tail(10)
```

## Renaming and removing units

```python
final_df = pd.DataFrame()
final_df['Cold Rent'] = df['Rent']
final_df['Rent incl. Extra in €'] = pd.to_numeric(df['Rent'].str.replace('€', '').replace('n.a.', np.nan), errors='coerce').fillna(0).astype(int) + pd.to_numeric(df['Extra_Costs'].str.replace('€', '').replace('n.a.', np.nan), errors='coerce').fillna(0).astype(int)
final_df['Deposit'] = df['Deposit'].str.replace('€', '').replace('n.a.', np.nan)
final_df['Roomsize in m²'] = df['Size'].str.replace('m²', '')
final_df['Appartmentsize in m²'] = df['Wohnungsgröße'].str.replace('Wohnungsgröße: ', '').str.replace('m²', '')
final_df['Flatmate count'] = df['WG_Size']
# PLZ from the address
plz_pattern = re.compile(r'\b\d{5}\b')
final_df['Postal Code'] = df['Address'].apply(lambda x: plz_pattern.search(x).group(0) if plz_pattern.search(x) else np.nan)

final_df['Floor'] = df['Geschoss'].str.replace('. OG', '').str.replace('Hochparterre', '0').str.replace('Tiefparterre', '0').replace('', np.nan).str.replace('EG', '0')
final_df['Online since in h'] = df['Online_Since'].apply(lambda x: 0 if 'Minuten' in x else int(x.split(' ')[0]) if 'Stunden' in x else int(x.split(' ')[0]) * 24 if ('Tage' in x or 'Tag' in x) else np.nan)

# Min and Max Age
def extract_min_age(row):
    # Extrahieren der Altersinformationen aus beiden Spalten
    ages_col1 = [int(age) for age in re.findall(r'\d+', str(row['Search_Info']))]
    ages_col2 = [int(age) for age in re.findall(r'\d+', str(row['Alter']))]
    
    # Kombinieren der Listen und Entfernen von leeren Werten
    all_ages = ages_col1 + ages_col2
    all_ages = [age for age in all_ages if age != []]
    
    # Rückgabe des Mindestalters, falls vorhanden, sonst NaN
    return (min(all_ages), max(all_ages)) if all_ages else np.nan
final_df[['Min Age', 'Max Age']] = df.apply(extract_min_age, axis=1).apply(pd.Series)

# Smoking
# smoking_df = pd.DataFrame()
# smoking_df['Rauchen'] = df['Rauchen'].str.replace('Rauchen im Zimmer erlaubt', 'Rauchen teilweise erlaubt').str.replace('Rauchen auf dem Balkon erlaubt', 'Rauchen teilweise erlaubt')
# smoking_df_dummies = pd.get_dummies(smoking_df, columns=['Rauchen'], drop_first=True) # Droping the first column to avoid multicollinearity (Rauchen nicht erlaubt)
# final_df = pd.concat([final_df, smoking_df_dummies], axis=1)
final_df['Smoking'] = df['Rauchen'].str.replace('Rauchen im Zimmer erlaubt', 'Rauchen teilweise erlaubt').str.replace('Rauchen auf dem Balkon erlaubt', 'Rauchen teilweise erlaubt')

# Parking
# parking_df_dummies = pd.get_dummies(df['Parken'], drop_first=True) # Droping the first column to avoid multicollinearity (Bewohnerparken)
# final_df = pd.concat([final_df, parking_df_dummies], axis=1)
final_df['Parking'] = df['Parken']

# Bautyp
# bautyp_df_dummies = pd.get_dummies(df['Bautyp'], drop_first=True) # Droping the first column to avoid multicollinearity (Altbau)
# final_df = pd.concat([final_df, bautyp_df_dummies], axis=1)
final_df['Building Type'] = df['Bautyp']

# Intermediate Rent
final_df['Intermediate Rent'] = df['Available_Till'].apply(lambda x: True if x!='n.a.' else False)
final_df.head(10)

# TODO: Online since überarbeiten --> irgendwann wird ein Datum eingegeben --> umwandeln und umrechnen in Stunden
```
